## 6. GroupBy & Aggregations
We can group by species and calculate statistics.

```python
iris.groupby('species')['sepal_length'].mean()
```

```python
iris.groupby('species')['sepal_length'].mean()
iris.groupby('species')['petal_length'].min()

iris.groupby('species').agg({'sepal_length':'mean', 'petal_length':'max'})
```

## 7. Merging & Joining
You can combine DataFrames using `merge` or `concat`.

```python
df1 = iris[['sepal_length','species']].head(5)
df2 = iris[['petal_length','species']].head(5)

merged = pd.merge(df1, df2, on='species')
merged
```

```python
df3 = iris[['sepal_length','petal_length']].head(5)
df4 = iris[['sepal_width','petal_width']].head(5)

concat = pd.concat([df3, df4], axis=1)
concat
```
